## SMOTE and Some Preprocessing

Defaulters (1) have a higher median bill-to-limit ratio than non-defaulters (0).
This means defaulters consistently use more of their credit limit.
The distribution for defaulters is more spread, showing that some users fluctuate wildly in billing.

```python
#Using SMOTE to get rid of imbalanced dataset
from imblearn.over_sampling import SMOTE
from sklearn.model_selection import train_test_split

# Step 1: Split the original data
X = train_df.drop(columns=['next_month_default'])
y = train_df['next_month_default']

X_train, X_test, y_train, y_test = train_test_split(X, y, stratify=y, test_size=0.2, random_state=42)

# Step 2: Apply SMOTE on training data
smote = SMOTE(random_state=42)
X_train_smote, y_train_smote = smote.fit_resample(X_train, y_train)

# Step 3: Reconstruct DataFrame from SMOTE output
columns = X.columns.tolist()
balance_df = pd.DataFrame(X_train_smote, columns=columns)
balance_df['next_month_default'] = y_train_smote

# Check new class balance
print("Before SMOTE:", y_train.value_counts())
print("After SMOTE:", y_train_smote.value_counts())
```

```python
y_train_resampled_df = pd.DataFrame({'next_month_default': y_train_smote})

sns.countplot(x='next_month_default', data=y_train_resampled_df, palette='Set1')
plt.title("Class Distribution After SMOTE")
plt.xlabel("Default (0 = No, 1 = Yes)")
plt.ylabel("Count")
plt.show()
```

```python
# 4.1 Total Payment Delay Score
balance_df['payment_value'] = balance_df[['pay_0', 'pay_2', 'pay_3', 'pay_4', 'pay_5', 'pay_6']].sum(axis=1)
```

```python
# 4.2 Total Dues = total billed - total paid
# Correcting the column names to use a capital 'B' for 'Bill_amt'
balance_df['dues'] = balance_df[[f'Bill_amt{i}' for i in range(1, 7)]].sum(axis=1) - \
                     balance_df[[f'pay_amt{i}' for i in range(1, 7)]].sum(axis=1)
```

```python
# 4.3 Clean education and marriage (no column renaming)
balance_df['education'] = np.where(balance_df['education'].isin([0, 5, 6]), 4, balance_df['education'])
balance_df['marriage'] = np.where(balance_df['marriage'] == 0, 3, balance_df['marriage'])
```

```python
balance_df['zero_pay_count'] = balance_df[[f'pay_amt{i}' for i in range(1, 7)]].apply(lambda row: (row == 0).sum(), axis=1)
```

```python
balance_df.head()
```

```python
# 1. One-hot encode EDUCATION and MARRIAGE
df_fr = pd.get_dummies(balance_df, columns=['education', 'marriage'])

# 2. Dynamically drop dummy trap columns (highest encoded category)
# Find all education columns and drop the last one
edu_cols = [col for col in df_fr.columns if col.startswith('education_')]
mar_cols = [col for col in df_fr.columns if col.startswith('marriage_')]

if edu_cols:
    df_fr.drop(edu_cols[-1], axis=1, inplace=True)
if mar_cols:
    df_fr.drop(mar_cols[-1], axis=1, inplace=True)

# 3. One-hot encode PAY columns (drop_first to avoid dummy trap)
df_fr = pd.get_dummies(df_fr, columns=[
    'pay_0', 'pay_2', 'pay_3', 'pay_4', 'pay_5', 'pay_6'
], drop_first=True)

# 4. Label encode SEX
df_fr['sex'] = df_fr['sex'].replace({'FEMALE': 0, 'MALE': 1})

# 5. Drop ID if it exists
if 'ID' in df_fr.columns:
    df_fr.drop('ID', axis=1, inplace=True)

# 6. Preview
df_fr.head()
```

```python
df_final = df_fr.copy()

df_final['avg_bill_amt'] = df_final[[f'Bill_amt{i}' for i in range(1, 7)]].mean(axis=1)
df_final['avg_pay_amt'] = df_final[[f'pay_amt{i}' for i in range(1, 7)]].mean(axis=1)
df_final['pay_to_bill_ratio'] = df_final['avg_pay_amt'] / (df_final['avg_bill_amt'] + 1)
df_final['pay_amt_std'] = df_final[[f'pay_amt{i}' for i in range(1, 7)]].std(axis=1)
```

```python
df_final['avg_pay_status'] = balance_df[['pay_0', 'pay_2', 'pay_3', 'pay_4', 'pay_5', 'pay_6']].mean(axis=1)
df_final['max_pay_status'] = balance_df[['pay_0', 'pay_2', 'pay_3', 'pay_4', 'pay_5', 'pay_6']].max(axis=1)
df_final['months_delayed'] = balance_df[['pay_0', 'pay_2', 'pay_3', 'pay_4', 'pay_5', 'pay_6']].apply(lambda row: (row >= 1).sum(), axis=1)
```

```python
if 'Customer_ID' in df_final.columns:
    df_final.drop('Customer_ID', axis=1, inplace=True)

df_final['dues_to_limit_ratio'] = df_final['dues'] / (df_final['LIMIT_BAL'] + 1)

from sklearn.linear_model import LinearRegression
def compute_payment_trend(row):
    months = np.arange(1, 7).reshape(-1, 1)
    model = LinearRegression().fit(months, row.values.reshape(-1, 1))
    return model.coef_[0][0]
```

```python
df_final['pay_trend'] = df_final[[f'pay_amt{i}' for i in range(1, 7)]].apply(compute_payment_trend, axis=1)
df_final['recent_late'] = (balance_df['pay_0'] >= 1).astype(int)
df_final['last_two_months_late'] = ((balance_df['pay_0'] >= 1) | (balance_df['pay_2'] >= 1)).astype(int)
df_final['high_usage_flag'] = (df_final['avg_bill_amt'] > 0.9 * df_final['LIMIT_BAL']).astype(int)
```

```python
df_final.head()
```

```python
X = df_final.drop('next_month_default', axis=1)
y = df_final['next_month_default']
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler

X_train, X_test, y_train, y_test = train_test_split(X, y, stratify=y, test_size=0.33, random_state=42)

scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

## MODEL TRAINING

```python
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from lightgbm import LGBMClassifier
from xgboost import XGBClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, fbeta_score, roc_auc_score, confusion_matrix
import seaborn as sns
```

```python
def evaluate_model(y_true, y_pred, y_proba, model_name):
    print(f"\n{model_name} Evaluation:")
    print("Accuracy:", accuracy_score(y_true, y_pred))
    print("Precision:", precision_score(y_true, y_pred))
    print("Recall:", recall_score(y_true, y_pred))
    print("F1 Score:", f1_score(y_true, y_pred))
    print("F2 Score:", fbeta_score(y_true, y_pred, beta=2))
    print("ROC AUC:", roc_auc_score(y_true, y_proba))

    sns.heatmap(confusion_matrix(y_true, y_pred), annot=True, fmt='d', cmap='Blues')
    plt.title(f"Confusion Matrix ({model_name})")
    plt.xlabel("Predicted")
    plt.ylabel("Actual")
    plt.show()
```

```python
# Logistic Regression
log_model = LogisticRegression(solver='liblinear')
log_model.fit(X_train_scaled, y_train)
log_pred = log_model.predict(X_test_scaled)
log_proba = log_model.predict_proba(X_test_scaled)[:, 1]
evaluate_model(y_test, log_pred, log_proba, "Logistic Regression")
```

```python
# Decision Tree
tree_model = DecisionTreeClassifier(max_depth=6, min_samples_leaf=5, random_state=42)
tree_model.fit(X_train_scaled, y_train)
tree_pred = tree_model.predict(X_test_scaled)
tree_proba = tree_model.predict_proba(X_test_scaled)[:, 1]
evaluate_model(y_test, tree_pred, tree_proba, "Decision Tree")
```

```python
# LightGBM
lgbm_model = LGBMClassifier(random_state=42)
lgbm_model.fit(X_train_scaled, y_train)
lgbm_pred = lgbm_model.predict(X_test_scaled)
lgbm_proba = lgbm_model.predict_proba(X_test_scaled)[:, 1]
evaluate_model(y_test, lgbm_pred, lgbm_proba, "LightGBM")
```

```python
# Random Forest
rf_model = RandomForestClassifier(n_estimators=100, random_state=42)
rf_model.fit(X_train_scaled, y_train)
rf_pred = rf_model.predict(X_test_scaled)
rf_proba = rf_model.predict_proba(X_test_scaled)[:, 1]
evaluate_model(y_test, rf_pred, rf_proba, "Random Forest")
```

```python
# XGBoost
xgb_model = XGBClassifier(use_label_encoder=False, eval_metric='logloss', random_state=42)
xgb_model.fit(X_train_scaled, y_train)
xgb_pred = xgb_model.predict(X_test_scaled)
xgb_proba = xgb_model.predict_proba(X_test_scaled)[:, 1]
evaluate_model(y_test, xgb_pred, xgb_proba, "XGBoost")
```
